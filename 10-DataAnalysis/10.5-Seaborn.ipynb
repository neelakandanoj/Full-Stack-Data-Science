```python
import seaborn as sns
tip=sns.load_dataset('tips')
print(tip)
```

```python
##Scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tip)
plt.title('Scatter plot of total bill vs tip')
```

```python
## Line plot

sns.lineplot(x='size',y='total_bill',data=tip)
plt.title('Line Plot of total bill vs size')
plt.show()
```

## Categorical plots

```python
tip.head()
```

```python
## BAR Plot

sns.barplot(x='day',y='total_bill',data=tip)
plt.title('Bar Plot for Total bill vs day')
plt.show()
```

## Box plot to check the outlier's

```python
## Box Plot

sns.boxplot(x='day',y='total_bill',data=tip)
```

```python
## Violin Plot

sns.violinplot(x='day',y='total_bill',data=tip)
```

```python
## Histogram

sns.histplot(tip['total_bill'],bins=10,kde=True)
```

```python
## KDE plot
sns.kdeplot(tip['total_bill'],fill='blue')
```

```python
## Pair Plot

sns.pairplot(tip)
```

```python
## heatMap

corr=tip[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True)
```

```python
import pandas as pd
sales_df=pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
# plot total sales by product

plt.figure(figsize=(10,6))
sns.barplot(x='Product Category',y='Total Revenue',data=sales_df,estimator=sum)
plt.title('Total sales by product')
plt.show()
```

```python
# plot total sales by Region

plt.figure(figsize=(10,6))
sns.barplot(x='Region',y='Total Revenue',data=sales_df,estimator=sum)
plt.title('Total sales by Region')
plt.show()
```
