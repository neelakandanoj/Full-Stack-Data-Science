```python
import matplotlib.pyplot as plt
```

## **Line Plot's**

```python
x=[1,2,3,4,5]
y=[1,4,9,11,9]
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('y axis')
plt.title('Basic line plot')
```

```python
x=[1,2,3,4,5]
y=[1,4,9,11,9]

#create customized line plot

plt.plot(x,y,color='red',linestyle='-.',marker='o',linewidth=2,markersize=9)
plt.grid(True)
```

```python
## Multiple plot
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(y1,x,color='blue')
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y2,color='red')
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(y1,x,color='yellow')
plt.title("Plot 4")
```

## **Bar Plot's**  -- x-axis Category values and y-axis Continuous values used to check distribution of category columns

```python
#Bar plot
categories=['A','B','C','D','E']
values=[5,7,3,8,6]

plt.bar(categories,values,color='purple')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar plot')
plt.show()
```

## **Histograms**  x-axis-Continuous values, y-axis- Categories values used to find distribution of continuous columns
